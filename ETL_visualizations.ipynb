```python
#Dependencies
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#CSV files read into pandas dataframes
Appstore_df = pd.read_csv("./HTML/Apple_table.csv")
Googleplay_df = pd.read_csv("./HTML/Google_table.csv")
```

```python
#Renaming the price column values as required
Appstore_df.loc[Appstore_df['price'] == 0.00, 'price'] = "Free"
Appstore_df.loc[Appstore_df['price'] != "Free", 'price'] = "Paid"
Appstore_df.head()
```

```python
#Renaming the price column values as required
Googleplay_df.loc[Googleplay_df['price'] == 0.0, 'price'] = "Free"
Googleplay_df.loc[Googleplay_df['price'] == 0.00, 'price'] = "Free"
Googleplay_df.loc[Googleplay_df['price'] != "Free", 'price'] = "Paid"
Googleplay_df.head()
```

```python
#Using the googleplay df, getting the app count for each genre
genre_grouped_Google = Googleplay_df.groupby(["genre"])["app_name"].count().sort_values(ascending=True)
#Plots horizontal bar chart with apps count per genre
genre_grouped_Google.plot(kind="barh",figsize=(10,8), fontsize=12)
plt.title("GooglePlay- No.of Apps by Genre",fontsize= 25, fontweight="bold")
plt.xlabel("Number Of Apps", fontsize= 20)
plt.ylabel("Genre", fontsize=20)
plt.savefig("./HTML/GooglePlay_Apps_by_Genre.png", bbox_inches="tight")
```

```python
#Using the appstore df, getting the app count for each genre
genre_grouped_Apple = Appstore_df.groupby(["genre"])["app_name"].count().sort_values(ascending=True)
#Plots horizontal bar chart with apps count per genre
genre_grouped_Apple.plot(kind="barh",figsize=(10,8), fontsize=10)
plt.title("AppStore- No.of Apps by Genre",fontsize= 25, fontweight="bold")
plt.xlabel("Number Of Apps", fontsize= 20)
plt.ylabel("Genre", fontsize=20)
plt.savefig("./HTML/AppStore_Apps_by_Genre.png", bbox_inches="tight")
```

```python
#Using the Googleplay df, getting the average user rating for each genre
genre_ratings_Google = Googleplay_df.groupby(["genre"])["user_rating"].mean().sort_values(ascending=True)
#Plots horizontal bar chart with apps count per genre
genre_ratings_Google.plot(kind="barh",figsize=(10,8), fontsize=12)
plt.title("GooglePlay- No.of Ratings by Genre",fontsize= 25, fontweight="bold")
plt.xlabel("Number Of Apps", fontsize= 20)
plt.ylabel("Genre", fontsize=20)
plt.savefig("./HTML/GooglePlay_Ratings_by_Genre.png", bbox_inches="tight")
```

```python
#Using the AppStore df, getting the average user rating for each genre
genre_ratings_Apple = Appstore_df.groupby(["genre"])["user_rating"].mean().sort_values(ascending=True)
#Plots horizontal bar chart with apps count per genre
genre_ratings_Apple.plot(kind="barh",figsize=(7,5), fontsize=8)
plt.title("AppStore- No.of Ratings by Genre",fontsize= 25, fontweight="bold")
plt.xlabel("Number Of Apps", fontsize= 20)
plt.ylabel("Genre", fontsize=20)
plt.savefig("./HTML/AppStore_Ratings_by_Genre.png", bbox_inches="tight")
```

```python
#Free and Paid apps analysis for AppStore dataset
Appstore_df.groupby("price")["price"].count()
```

```python
#Free and Paid apps analysis for Googleplay dataset
Googleplay_df.groupby("price")["price"].count()
```

```python
#PIE chart for paid and unpaid apps - APPSTORE
colors = ["#3ccedf","#ff3208"]
Appstore_df.price.value_counts().plot(kind='pie', autopct='%1.0f%%', pctdistance=0.5,shadow=True,legend=True,colors=colors)
plt.title("APPSTORE - % of Free/Paid Apps")
plt.savefig("AppStore-%Free_Paid_.png")
#The purpose of this figure is to show the AppStore's and PlayStore's percentages of free and paid category applications for the year 2017. 
We could see that for "Google PlayStore", there is a huge difference between the percentages of free and paid apps whereas for "Apple AppStore", we could only see a slight difference of less than five percent.
This clearly says that Google managed to maintained a higher ratio of free apps than paid apps but, Apple did not show any considerable difference of ratios between paid and free apps.
                                            
```

```python
#PIE chart for paid and unpaid apps - GooglePlay
colors = ["#4ede37","#ff3208"]
Googleplay_df.price.value_counts().plot(kind='pie', autopct='%1.0f%%', pctdistance=0.5,shadow=True,legend=True,colors=colors)
plt.title("GOOGLEPLAY - % of Free/Paid Apps")
plt.savefig("GooglePlay-%Free_Paid_.png")
```
